```python
from datasets import load_dataset
import pandas as pd
import os

dataset = load_dataset("abisee/cnn_dailymail", "3.0.0")

os.makedirs('data', exist_ok=True)

pd.DataFrame(dataset["train"]).to_csv("data/cnn_train.csv", index=False)
pd.DataFrame(dataset["validation"]).to_csv("data/cnn_validation.csv", index=False)
pd.DataFrame(dataset["test"]).to_csv("data/cnn_test.csv", index=False)
```
